## **1.1 Latar Belakang**

Mengelola risiko portofolio asuransi membutuhkan ketelitian dan kehati-hatian yang tinggi, dalam mengidentifikasi risiko tertanggung yang menyebabkan pengklaiman polis asuransi sesuai cakupan yang ada didalam polis. Namun proses ini bisa memakan waktu yang lama. Banyaknya klaim yang diajukan juga dapat menyebabkan kerugian finansial langsung bagi perusahaan asuransi, karena mereka harus membayar klaim sesuai dengan cakupan polis. Tingginya jumlah klaim dapat menyebabkan kenaikan premi di masa depan, yang dapat mempengaruhi daya saing produk asuransi dan kepuasan pelanggan.

Apabila perusahaan dapat memprediksi profil risiko calon nasabah, maka perusahaan dapat menetapkan harga premi yang lebih sesuai. Di sisi lain dengan mengetahui calon nasabah yang kemungkinan akan mengajukan klaim, perusahaan dapat menawarkan layanan tambahan atau dukungan proaktif untuk mengurangi kemungkinan terjadinya klaim atau membantu pelanggan dalam situasi darurat. Dengan mengurangi jumlah klaim yang tidak terduga, perusahaan dapat mengurangi biaya yang dikeluarkan untuk membayar klaim, sehingga meningkatkan profitabilitas.

Selain itu, dengan mengetahui kemungkinan pelanggan akan mengklaim polis, perusahaan dapat lebih akurat memperkirakan cadangan klaim yang diperlukan, sehingga mengelola keuangan perusahaan dengan lebih baik. Jadi, dengan memprediksi klaim, perusahaan dapat mengidentifikasi pelanggan atau jenis polis yang memiliki risiko klaim lebih tinggi. Ini memungkinkan perusahaan untuk mengambil tindakan pencegahan yang lebih baik dan mengelola portofolio risiko mereka lebih efektif.

## **1.2 Pernyataan Masalah**

Berdasarkan latar belakang tersebut, **perusahaan ingin memiliki kemampuan mendeteksi klaim yang sebenarnya terjadi, sehingga dapat mengurangi risiko kerugian finansial karena klaim yang tidak terdeteksi.**

## **1.3 Tujuan**

2. Menyesuaikan harga premi berdasarkan profil risiko yang diprediksi.
    - Menjaga produk tetap kompetitif.
    - Meningkatkan manajemen risiko.

## **1.4 Pendekatan Analisis**

Analisis dimulai dengan mengidentifikasi relevansi setiap fitur terhadap pernyataan masalah, mengecek korelasi antara dua fitur untuk memahami bagaimana satu fitur mempengaruhi satu fitur lain. Menentukan fitur penting yang berpengaruh terhadap variabel target yaitu klaim polis asuransi, dilakukan dengan pertimbangan menggunakan pengetahuan domain.Fitur yang dipilih selanjutnya dilakukan praproses. Setelah fitur sudah siap selanjutnya membuat model klasifikasi yang dapat membantu perusahaan untuk memprediksi probabilitas seorang tertanggung mengklaim polis asuransi.

## **1.5 Metrik Evaluasi**

Pertimbangan konsekuensi tipe kesalahan prediksi model terhadap bisnis sebagai berikut:

**Kesalahan tipe 1 _False Positive_:** Salah memprediksi tertanggung akan mengklaim asuransi padahal sebenarnya tidak.

**Konsekuensi:** Seorang Underwriter akan salah sasaran menentukan harga premi sehingga premi yang terlalu tinggi dapat membuat produk asuransi tidak kompetitif, sehingga calon nasabah mungkin memilih perusahaan asuransi lain yang menawarkan premi lebih rendah. Penjualan yang rendah karena premi yang mahal dapat mengurangi pendapatan perusahaan.

**Kesalahan tipe 2 _False Negative_:** Salah memprediksi tertanggung tidak akan mengklaim asuransi padahal sebenarnya mengklaim.

**Konsekuensi:** Pendapatan premi yang lebih rendah daripada klaim yang harus dibayarkan perusahaan dapat menurunkan profitabilitas perusahaan. Dalam jangka panjang, jika banyak klaim diajukan dan simpanan keuangan perusahaan tidak mencukupi, perusahaan bisa menghadapi masalah keuangan.

Berdasarkan konsekuensinya, maka sebisa mungkin model dapat dengan baik memprediksi probabilitas kelas positif yang memang nasabah mengklaim asuransi. Oleh karena itu **metrik yang digunakan adalah _Recall_.** Yaitu proporsi aktual positif yang terdeteksi oleh model.

#### 4.2.3.a: Proporsi Klaim Asuransi Antara Pria dan Wanita

```python
male_claim_yes = len(df[(df['gender'] == 'M') & (df['claim'] == 'Yes')])
female_claim_yes = len(df[(df['gender'] == 'F') & (df['claim'] == 'Yes')])

male_claim_no = len(df[(df['gender'] == 'M') & (df['claim'] == 'No')])
female_claim_no = len(df[(df['gender'] == 'F') & (df['claim'] == 'No')])

claims_yes = pd.DataFrame({'Gender': ['Female', 'Male'], 'Claims': [female_claim_yes, male_claim_yes]})
claims_no = pd.DataFrame({'Gender': ['Female', 'Male'], 'Claims': [female_claim_no, male_claim_no]})

fig, axes = plt.subplots(1, 2, figsize=(10,5))
sns.barplot(ax=axes[0], x='Gender', y='Claims', data=claims_yes, palette=['red', 'blue'])
axes[0].set_title('Jumlah Klaim "Yes" Berdasarkan Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Jumlah Klaim')

sns.barplot(ax=axes[1], x='Gender', y='Claims', data=claims_no, palette=['red', 'blue'])
axes[1].set_title('Jumlah Klaim "No" Berdasarkan Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Jumlah Klaim')

plt.tight_layout()
plt.show()

print('Klaim Yes')
print(claims_yes, '\n')
print('Klaim No')
print(claims_no)
```

Dapat disimpulkan secara jumlah asuransi yang diklaim dan tidak pada antara pria dan wanita hanya sedikit perbedaannya, tetapi apakah kesimpulan ini valid atau hanya kebetulan semata, kita perlu menguji hipotesis kita menggunakan metode pengujian statistik Chi-Square untuk menentukan apakah ada hubungan atau asosiasi yang signifikan antara dua variabel kategorikal, dalam hal ini variabel adalah gender pria dan wanita.

Hipotesis kita adalah:
- H0: Tidak ada hubungan yang signifikan antara gender dan klaim asuransi perjalanan.
- Ha: Ada hubungan yang signifikan antara gender dan klaim asuransi perjalanan.

```python
claim_distribution = df.groupby('gender')['claim'].value_counts(normalize=True).unstack()
claim_stats = df.groupby('gender')['claim'].apply(lambda x: (x == 'Yes').mean())
contingency_table = pd.crosstab(df['gender'], df['claim'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print('Proporsi klaim asuransi antara pria dan wanita')
print(claim_distribution, '\n')
print('Rata-rata proporsi klaim (Yes) antara pria dan wanita')
print(claim_stats, '\n')
print('Uji Chi-Square')
if p >= 0.05:
    print('p-value', p, '\n')
    print('H0 diterima')
    print('GenderÂ danÂ klaimÂ asuransiÂ perjalananÂ tidakÂ terkait')
else:
    print('p-value', p, '\n')
    print('H0 ditolak, kita memiliki cukup bukti untuk menerima Ha')
    print('GenderÂ danÂ klaimÂ asuransiÂ perjalananÂ terkait')
```

Dari hasil uji hipotesis kita dapat diambil kesimpulan bahwa tidak ada perbedaan dalam proporsi klaim antara pria dan wanita. Artinya, peluang untuk mengajukan klaim asuransi perjalanan sama untuk kedua gender.

## **4.4: Praproses**

Setelah kita selesai memilih fitur, membersihkan data, dan mensederhanakan fitur, kita dapat lanjut ke tahap mentransformasi data agar siap dimodelkan. Berikut adalah skema praproses _(preprocessing)_ yang dilakukan:
1. One-hot encoding: _Agency Type, Gender_
2. Binary encoding: _Product Name, Destination_
3. Ordinal encoding: fitur _Age_ setelah binning

**Penjelasan lebih lanjut praproses yang dilakukan**

Pengkodean pada fitur kategorik dilakukan karena algoritma yang akan digunakan pada proyek ini membutuhkan input numerik untuk melakukan perhitungan matematis. Data kategorik dalam bentuk _string_ atau label tidak dapat langsung digunakan dalam operasi matematis tersebut. Jadi, dengan mengonversi data kategorik menjadi bentuk numerik, model dapat mempelajari pola dari data. Representasi numerik memungkinkan algoritma untuk menghitung jarak, kemiripan, dan korelasi antara fitur yang berbeda.

Teknik pengkodean yang kita gunakan ada tiga, pertama _one-hot encoding_ yang diterapkan ke fitur _Agency Type,_ dan _Gender_. Ketiga fitur ini adalah data nominal, jadi setiap kategori direpresentasikan sebagai vektor biner yang memungkinkan model untuk membedakan antara kategori yang berbeda.

Kedua, _binary encoding_ yang diterapkan ke fitur _Product Name,_ dan _Destination_, ketiga fitur ini memiliki data kategori yang banyak dan tidak memiliki ciri eksklusif atau urutan. Jadi, setiap kategori diubah menjadi angka kemudian dikonversi menjadi biner. Setiap digit biner kemudian dipecah menjadi kolom terpisah. Teknik pengkodean ini berguna untuk data yang memiliki kategori banyak tapi dapat lebih efisien karena representasi biner mengurangi dimensi.

Ketiga, _ordinal encoding_ yang diterapkan ke fitur _Age_ yang sudah di _binning_ akan menjadi data kategori yang memiliki urutan, dimana dalam konteks bisnis ini data usia dikelompokan kedalam kategori wisatawan berusia lanjut _Senior Citizen Travelers_ yang memiliki risiko paling tinggi, sehingga wisatawan ini masuk dalam urutan pertama. Lalu wisatawan berusia dewasa _Adult Travelers_ memiliki risiko tertinggi kedua, wisatawan dewasa muda _Young Adult Travelers_ memiliki risiko tertinggi ketiga, wisatawan anak-anak _Child Travelers_ memiliki risiko tertinggi keempat.

Urutan keempat kelompok wisatawan ini telah dipertimbangkan dari sisi kesehatan dan frekuensi klaim. Namun perlu diingat untuk kejadian tidak terduga lainnya yang dilindungi dalam asuransi, keempat kelompok wisatawan ini memiliki tingkat risiko yang sama. Jadi, pengkodean ordinal akan mengubah kategori menjadi angka yang mencerminkan urutan kategori tersebut.

```python
ordinal_mapping = [{'col': 'age_group',
                    'mapping':{'Senior Citizen Travelers': 1, 'Adult Travelers': 2, 'Young Adult Travelers': 3, 'Child Travelers':4}}
                  ]

binary_encoder = ce.BinaryEncoder()
ordinal_encoder = ce.OrdinalEncoder(mapping=ordinal_mapping)

preprocessor = ColumnTransformer(transformers=[
    ('one_hot_encoding', OneHotEncoder(drop='first'), ['agency_type', 'gender']),
    ('binary_encoding', binary_encoder, ['product_name', 'destination']),
    ('ordinal_encoding', ordinal_encoder, ['age_group'])
], remainder='passthrough')
```

Kode pada cell diatas adalah kerangka pemrosesan data yang diurutkan ke dalam fungsi ColumnTransformer sehingga bisa kita gunakan kapanpun di tahap selanjutnya.

Pada tahap ini kita latih beberapa model dasar dengan parameter _default_ untuk mendapatkan _baseline performance_. Secara garis besar langkah-langkah yang akan kita terapkan terdiri dari:

1. Mengubah kolom target menjadi representasi numerik, 1 untuk _'Yes'_ dan 0 untuk _'No'_, sehingga kita bisa latih ke model.
2. Pemisahan data _(data splitting)_ target dari fitur, kita akan membagi data 80% untuk dilatih model, kemudian hasil model akan diuji ke 20% data tes. Untuk memastikan proporsi target sama antara data latih dan uji, kita akan menstratifikasi kolom target.

5. Setelah kita mendapat satu model _benchmark_, kita gunakan model itu seterusnya. Kita akan tingkatkan kinerja model menggunakan beberapa teknik, diantaranya membuat fitur baru dan melakukan _resampling_ untuk mengimbangi data target.

- **K-Nearest Neighbors (KNN)** digunakan untuk menangkap hubungan non-linear dalam data, dengan keunggulannya dalam kesederhanaan dan kemampuan beradaptasi pada berbagai distribusi data.

Proses training dari logistic regression melibatkan penyesuaian koefisien model untuk meminimalkan kesalahan antara output prediksi dan label aktual. Logistic Regression memberikan bobot pada setiap fitur input, dan bobot ini menunjukkan seberapa kuat masing-masing fitur mempengaruhi probabilitas hasil.

Kelebihan dari logistic regression meliputi interpretasi yang mudah, di mana koefisien dapat membantu memahami pengaruh masing-masing fitur terhadap hasil. Model ini juga efektif dalam menangani kasus klasifikasi biner yang linear.

Namun kelemahan dan batasan dari logistic regression adalah, model ini mengasumsikan hubungan linear antara fitur input dan log-odds dari hasil, yang berarti tidak bekerja dengan baik jika hubungan antara fitur dan target adalah non-linear. Logistic Regression juga rentan terhadap outliers, yang dapat mempengaruhi estimasi koefisien. Logistic Regression juga tidak dapat secara otomatis menangani interaksi antara fitur, yang dapat membatasi performanya dalam dataset yang kompleks. Oleh karena itu, ketika menggunakan logistic regression, sangat penting untuk memastikan bahwa data telah dipra-proses dengan baik dan fitur-fitur yang relevan telah dipilih.

### **5.3.1: Fitur Polynomial**

_Polynomial features_ adalah teknik yang digunakan untuk memperkaya representasi data dengan membuat fitur baru yang berasal dari kombinasi fitur asli yang dipangkatkan. Dalam konteks _machine learning_, terutama untuk model yang menangani hubungan linear seperti regresi linear dan logistic regression, menambahkan polynomial features dapat membantu model menangkap hubungan non-linear yang mungkin ada dalam data. Misalnya, jika dataset memiliki fitur _x1_ dan _x2_, _polynomial feature engineering_ dapat menghasilkan fitur baru seperti _x1^2_, _x2^2_, dan _x1_ * _x2_.

Teknik ini meningkatkan kompleksitas model dengan memungkinkan interaksi antar fitur, yang dapat meningkatkan kinerja model dalam memprediksi variabel target ketika ada pola non-linear. Meskipun menambahkan _polynomial features_ dapat meningkatkan akurasi model, terutama jika hubungan antara fitur dan target tidak linier, ini juga dapat meningkatkan risiko _overfitting_, terutama jika derajat polinomial yang digunakan terlalu tinggi atau jumlah fitur asli sangat besar. Oleh karena itu, penting untuk melakukan _feature selection_ dan _cross-validation_ secara hati-hati ketika menggunakan _polynomial features_ untuk memastikan bahwa model tidak menjadi terlalu rumit dan tetap dapat digeneralisasi dengan baik ke data baru.

Untuk menemukan derajat polinomial terbaik, kita akan melakukan perulangan polinomial dengan menggunakan derajat yang berbeda, lalu melatih dan langsung mengujinya menggunakan model logistic regression. Lalu, kita akan bandingkan diantara derajat polinomial yang ditentukan, mana yang menghasilkan kinerja model paling tinggi.

**Kita akan menggunakan dua fitur untuk diterapkan polinomial, yaitu fitur _Duration_ dan _Net Sales_, karena kedua fitur ini adalah fitur numerik kontinu.**

```python
degree = [2, 3, 4, 5]

best_degree = None
best_score_train = 0
best_score_test = 0
best_X_train_poly = None
best_X_test_poly = None

for i in degree:
  poly = PolynomialFeatures(degree=i, interaction_only=True)
  X_train_poly = poly.fit_transform(X_train_encoded[['remainder__duration', 'remainder__net_sales_SGD']])
  X_test_poly = poly.transform(X_test_encoded[['remainder__duration', 'remainder__net_sales_SGD']])

  poly_feature_names = poly.get_feature_names_out(['remainder__duration', 'remainder__net_sales_SGD'])
  X_train_poly = pd.DataFrame(X_train_poly, columns=poly_feature_names)
  X_test_poly = pd.DataFrame(X_test_poly, columns=poly_feature_names)

  X_train_poly = pd.concat([X_train_encoded, X_train_poly], axis=1)
  X_test_poly = pd.concat([X_test_encoded, X_test_poly], axis=1)

  logreg_model.fit(X_train_poly, y_train)

  y_pred_train = logreg_model.predict(X_train_poly)
  y_pred_test = logreg_model.predict(X_test_poly)

  score_train = recall_score(y_train, y_pred_train)
  score_test = recall_score(y_test, y_pred_test)

  print(f'Skor recall pada data train setelah menerapkan teknik Polynomial dengan deggree {i}:', score_train)
  print(f'Skor recall pada data test setelah menerapkan teknik Polynomial dengan deggree {i}:', score_test, '\n')

  if score_train + score_test > best_score_train + best_score_test:
        best_score_train = score_train
        best_score_test = score_test
        best_X_train_poly = X_train_poly
        best_X_test_poly = X_test_poly
        best_degree = i

print(f'Degree terbaik: {best_degree}')
print(f'Skor recall terbaik pada data train: {best_score_train}')
print(f'Skor recall terbaik pada data test: {best_score_test}')

X_train_poly = best_X_train_poly
X_test_poly = best_X_test_poly
```

Dapat dilihat outputnya bahwa derajat polinomial yang menghasilkan kinerja model terbaik adalah 2. Hasilnya juga bisa dilihat pada tabel diatas, fitur baru yang dihasilkan mencakup semua kombinasi kuadrat dan interaksi dari fitur asli tersebut. Dengan kata lain, teknik ini membuat transformasi dari fitur asli ke dalam bentuk polinomial derajat kedua. Jadi, untuk dua fitur input _x1_ dan _x2_, yaitu **remainder__duration** dan **remainder__net_sales_SGD**, fitur polinomial yang dihasilkan adalah:
1.	_x1:_ Fitur asli **remainder__duration**.
2.	_x2:_ Fitur asli **remainder__net_sales_SGD**.
3.	_x1^2:_ Kuadrat dari **remainder__duration**.
4.	_x2^2:_ Kuadrat dari **remainder__net_sales_SGD**.
5.	_x1_ * _x2:_ Interaksi antara **remainder__duration** dan **remainder__net_sales_SGD**.

Setelah menerapkan polynomial features dengan degree 2, dataset awal dengan dua fitur ini diperluas menjadi lima fitur. Penambahan fitur-fitur ini memungkinkan model untuk menangkap hubungan non-linear antara **remainder__duration** dan **remainder__net_sales_SGD** yang mungkin tidak terlihat dengan menggunakan fitur asli secara langsung.

Sebagai contoh, **remainder__duration** dan **remainder__net_sales_SGD**, fitur _x1^2_ dapat membantu mengidentifikasi pola di mana durasi memiliki efek kuadratik terhadap variabel target, sedangkan fitur interaksi _x1_ * _x2_ dapat menunjukkan bagaimana kombinasi dari durasi dan penjualan bersih mempengaruhi hasil.

## **5.4: Menangani Data Tidak Seimbang**

Seperti yang telah kita ketahui diawal bahwa data target sangat tidak berimbang, perbandingannya adalah 100:10.000 data target 1 dan data target 0.
Karena rasio yang sangat jauh ini kita akan menggunakan metode _resampling_ yaitu kombinasi dari _oversampling_ dan _undersampling_. Kita akan menguji 3 teknik kombinasi _resampling_ yaitu _Synthetic Minority Over-sampling Technique_ (SMOTE), gabungan SMOTE dan metode _undersampling: edited nearest neighbours_ (SMOTEENN), dan _Adaptive Synthetic Sampling Approach_ (ADASYN). Kemudian membandingkan hasil skor _recall_ pada masing-masing teknik _resampling_.

```python
logistic_regression = LogisticRegression(random_state=42)
```

```python
smote = SMOTE(random_state=42, n_jobs=-1)
smoteenn = SMOTEENN(random_state=42, n_jobs=-1)
adasyn = ADASYN(random_state=42, n_jobs=-1)

resampling = [smote, smoteenn, adasyn]

best_resampler = None
best_score_train = 0
best_score_test = 0
best_X_train_resampled = None
best_y_train_resampled = None

for resampler in resampling:
    X_train_resampled, y_train_resampled = resampler.fit_resample(X_train_poly, y_train)

    logistic_regression.fit(X_train_resampled, y_train_resampled)

    y_pred_train = logistic_regression.predict(X_train_resampled)
    y_pred_test = logistic_regression.predict(X_test_poly)

    score_train = recall_score(y_train_resampled, y_pred_train)
    score_test = recall_score(y_test, y_pred_test)

    print(f'Skor recall pada data train setelah teknik resampling {resampler.__class__.__name__}:', score_train)
    print('Confusion Matrix data latih:')
    print(confusion_matrix(y_train_resampled, y_pred_train), '\n')

    print(f'Skor recall pada data test setelah teknik resampling {resampler.__class__.__name__}:', score_test)
    print('Confusion Matrix data test:')
    print(confusion_matrix(y_test, y_pred_test), '\n')

    if score_train + score_test > best_score_train + best_score_test:
        best_score_train = score_train
        best_score_test = score_test
        best_resampler = resampler
        best_X_train_resampled = X_train_resampled
        best_y_train_resampled = y_train_resampled

print(f'Teknik resampling terbaik: {best_resampler.__class__.__name__}')
print(f'Skor recall terbaik pada data train: {best_score_train}')
print(f'Skor recall terbaik pada data test: {best_score_test}')

X_train_resampled = best_X_train_resampled
y_train_resampled = best_y_train_resampled
```

Dapat dilihat output diatas, teknik penanganan _imbalance data_ yang menghasilkan kinerja model lebih tinggi adalah SMOTEENN, yang merupakan gabungan SMOTE dan ENN.

SMOTE bekerja dengan membuat sampel sintetis dari kelas minoritas. Teknik ini menggunakan algoritma interpolasi untuk membuat titik data baru di antara sampel minoritas yang ada. Proses ini dilakukan dengan memilih titik-titik data minoritas yang ada dan kemudian membuat titik-titik baru di sepanjang garis yang menghubungkan titik-titik data yang dipilih dengan tetangga terdekat mereka. Dengan demikian, SMOTE meningkatkan jumlah sampel dari kelas minoritas, yang membantu dalam pelatihan model dengan lebih seimbang.

Setelah proses SMOTE, ENN diterapkan untuk membersihkan data yang dihasilkan. ENN bekerja dengan menghapus sampel yang salah klasifikasi atau tidak konsisten dari data. Teknik ini memeriksa setiap sampel dalam dataset yang dihasilkan (baik dari kelas mayoritas maupun minoritas) dan menghapusnya jika tidak memiliki mayoritas tetangga terdekat yang berasal dari kelas yang sama. ENN bertujuan untuk mengurangi noise dan menghilangkan sampel yang sulit, sehingga meningkatkan kualitas dataset dengan menghapus poin-poin yang dapat membingungkan model.

Dengan menggabungkan SMOTE dan ENN, SMOTEENN berusaha untuk tidak hanya meningkatkan jumlah sampel dari kelas minoritas tetapi juga untuk meningkatkan kualitas data dengan menghapus sampel yang tidak konsisten atau mengganggu. Ini menghasilkan dataset yang lebih seimbang dan bersih, yang dapat membantu model machine learning mempelajari pola yang lebih akurat dan menghindari bias terhadap kelas mayoritas.

## **6.1: Hyperparameter Tuning**

Kita akan menggunakan empat parameter utama didalam logistic regression, yaitu:
- penalty: Regularisasi yang digunakan dalam model.
- C: Inverse dari kekuatan regularisasi.
- solver: Algoritma yang digunakan dalam optimasi error.
- max_iter: Jumlah maksimum iterasi yang diambil oleh solver untuk konvergensi.

Dari keempat parameter tersebut beserta daftar nilainya akan menghasilkan 1200 kombinasi parameter. Tetapi, karena keterbatasan spesifikasi komputer milik program ini dijalankan, dan tidak efisiennya waktu, kita gunakan `RandomSearchCV` alih-alih menggunakan `GridSearch`, yang mencoba satu persatu dari semua kombinasi parameter. Tentu saja teknik grid akan memakan waktu sangat lama dan besarnya biaya lainnya. Maka dari itu, kita akan **secara acak memilih 10% dari total kombinasi parameter, yaitu 120 kombinasi**. Artinya teknik `RandomSearchCV` akan dijalankan berulang sebanyak 120 kali sesuai daftar kombinasi parameter yang dipilih secara acak.

```python
polynomial = PolynomialFeatures(degree=2, interaction_only=True)

transformer = ColumnTransformer(
    transformers=[
        ('one_hot_encoding', OneHotEncoder(drop='first'), ['agency_type', 'gender']),
        ('binary_encoding', binary_encoder, ['product_name', 'destination']),
        ('ordinal_encoding', ordinal_encoder, ['age_group']),
        ('feature_engineering', Pipeline(steps=[
            ('polynomial', polynomial)
        ]), ['duration', 'net_sales_SGD'])
    ],
    remainder='passthrough'
)

X_train_transformed = transformer.fit_transform(X_train)

smoteenn = SMOTEENN(random_state=42, n_jobs=-1)
X_train_resampled, y_train_resampled = smoteenn.fit_resample(X_train_transformed, y_train)

logreg_final_model = LogisticRegression(random_state=42, n_jobs=-1)

pipeline_training = Pipeline([
    ('model', logreg_final_model)
])

hyperparam_space = {
    'model__penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'model__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'model__max_iter': [100, 200, 300, 500, 1000],
}

random_search = RandomizedSearchCV(
    pipeline_training,
    param_distributions=hyperparam_space,
    n_iter=120,
    cv=skfold,
    scoring='recall',
    random_state=42,
    n_jobs=-1,
)
```

```python
random_search.fit(X_train_resampled, y_train_resampled)
```

```python
print('Parameter terbaik:', random_search.best_params_, '\n')
print('Skor recall tertinggi:', random_search.best_score_)
```
